## 2. Dataset Information

The dataset consists of 607 salary records of data science professionals from various experience levels, job roles, and geographic locations, spanning from 2020 to 2022.

Each record includes:
- Experience Level: Entry-Level (EN), Mid-Level (MI), Senior-Level (SE), and Executive-Level (EX).
- Employment Type: Full-Time (FT), Contract (CT), Freelance (FL), and Part-Time (PT).
- Job Titles: 50 unique data science-related job roles.
- Salary Information: Salaries recorded in different currencies but standardized to USD.
- Remote Work Ratio: On-Site (0%), Hybrid (50%), and Fully Remote (100%).
- Company Location: 50 different countries.
- Company Size: Small (S), Medium (M), Large (L).

This dataset provides an opportunity to analyze how salaries are influenced by experience, job role, company size, and work environment.


**Dataset Source: (Kaggle)** https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries

```python
# Import necessary libraries for data processing and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pycountry

file_path = "/Users/yash/Downloads/ds_salaries.csv"
df = pd.read_csv(file_path)

df = df.drop(columns=['Unnamed: 0'])


sns.set(style="whitegrid")

#Salary Trends Over Time
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="work_year", y="salary_in_usd", ci=None, marker='o', estimator='mean')
plt.xlabel("Year")
plt.ylabel("Average Salary (USD)")
plt.title("Average Data Science Salary Over Time")
plt.show()
```

### Explanation
A line plot showing average salary over years (work_years vs. salary_in_usd). it helps tracks salary growth trends in the data science field.

**Observation**:
- The dataset reveals a consistent increase in salaries from 2020 to 2022.
- The most significant jump in salaries occurred in 2021, possibly due to the growing demand for data professionals during the digital transformation driven by the COVID-19 pandemic.
- Entry-Level salaries remained relatively stable, while Senior-Level and Executive-Level salaries showed substantial growth over the years.

 **Insight**:
- The steady salary increase suggests that data science remains a lucrative and growing field.
- The 2021 salary spike may indicate that companies competed more aggressively for skilled professionals, leading to higher pay offers.
- Experience plays a crucial role in salary growth, as higher seniority levels benefit more from pay increases over time.


```python
# Salary Distribution by Experience Level
plt.figure(figsize=(8, 5))

# Generate visualizations to analyze and interpret data trends
sns.boxplot(x="experience_level", y="salary_in_usd", data=df, palette="muted")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.title("Salary Distribution by Experience Level")
plt.show()
```

### Explanation
The above bar plot or boxplot comparing salaries for different experience-level categories. It is useful to see how experience impacts salary.

 **Observation**:

- Executive-Level (EX) professionals earn the highest salaries, followed by Senior-Level (SE) and Mid-Level (MI) professionals.

 **Insight**:
-	The salary increase from Entry-Level to Senior-Level is almost double, emphasizing the importance of career progression.

```python
#  Impact of Remote Work on Salary
plt.figure(figsize=(8, 5))

# Generate visualizations to analyze and interpret data trends
sns.boxplot(x="remote_ratio", y="salary_in_usd", data=df, palette="coolwarm")
plt.xlabel("Remote Work Ratio (%)")
plt.ylabel("Salary in USD")
plt.title("Salary Distribution by Remote Work Percentage")
plt.show()
```

### Explanation
The above boxplot comparing remote_work_ratio with salaries_in_usd is shown, it helps check if remote jobs pay more or less.

**Observation**:

**Insight**:

```python
# Function to convert country codes
def convert_country_code(code):
    try:
        return pycountry.countries.lookup(code).alpha_3  # Convert ISO-2 to ISO-3
    except LookupError:
        return None

# Convert country codes
df["employee_residence"] = df["employee_residence"].apply(convert_country_code)

# Drop missing values after conversion
df = df.dropna(subset=["employee_residence"])

# Aggregate average salary per country
choropleth_data = df.groupby("employee_residence", as_index=False)["salary_in_usd"].mean()

# Create choropleth map
fig = px.choropleth(
    choropleth_data,
    locations="employee_residence",
    locationmode="ISO-3",
    color="salary_in_usd",
    title="Average Salary by Country (Employee Residence)",
    color_continuous_scale="viridis",
    labels={"salary_in_usd": "Avg Salary (USD)"}
)

fig.show()
```

### Explanation
The above choropleth map visualizes the average salary by country (employee residence). The color gradient ranges from dark purple (lower salaries) to bright yellow (higher salaries), representing salary distributions worldwide.

***Observation:***

United States, Canada, and Australia show relatively higher average salaries (green to yellow shades), indicating strong pay scales in these regions.
Western Europe (e.g., Germany, UK) also shows competitive salaries, though slightly lower than top-paying countries.
Asian countries, particularly India and some Middle Eastern nations, exhibit mid-range salaries (blue-green shades).
South American and African countries predominantly feature lower average salaries (darker purple shades).

***Insight:***

North America and parts of Western Europe appear to offer the highest earning potential for data-related roles.
Countries with emerging tech markets, like India and Brazil, demonstrate moderate pay, possibly reflecting growing industry presence.
Targeting roles in the United States, Australia, or Western Europe may offer higher salary prospects for data professionals seeking global opportunities.
